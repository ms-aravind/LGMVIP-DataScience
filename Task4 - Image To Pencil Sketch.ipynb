## Importing Modules

```python
import matplotlib.pyplot as mp
import cv2
mp.style.use('seaborn')
```

# Loading and Plotting Original Image

```python
img = cv2.imread('C:/Users/user/Desktop/LGM VIP/image.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
mp.figure(figsize=(8,8))
mp.imshow(img)
mp.axis("off")
mp.title("Original Image")
mp.show()
```

# Converting to Grayscale

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
mp.figure(figsize=(8,8))
mp.imshow(gray_img,cmap="gray")
mp.axis("off")
mp.title("Grayscale Image")
mp.show()
```

# Image Inverting

```python
invert_img = cv2.bitwise_not(gray_img)
mp.figure(figsize=(8,8))
mp.imshow(invert_img,cmap="gray")
mp.axis("off")
mp.title("Inverted Image")
mp.show()
```

 # Smoothing The Image To Make Sure The Sketch Is
#  LESS EDGY AND SMOOTH

```python
smooth_img = cv2.GaussianBlur(invert_img, (21, 21),sigmaX=0, sigmaY=0)
mp.figure(figsize=(8,8))
mp.imshow(smooth_img,cmap="gray")
mp.axis("off")
mp.title("Smooth Image")
mp.show()
```

# Converting image to Pencil Sketch

```python
sketch = cv2.divide(gray_img, 255 - smooth_img, scale=255)
mp.figure(figsize=(8,8))
mp.imshow(sketch,cmap="gray")
mp.axis("off")
mp.title("Final Image")
mp.show()
```

# Final Pencil Sketch 

```python
mp.figure(figsize=(20,20))
mp.subplot(1,2,1)
mp.imshow(img)
mp.axis("off")
mp.title("Original Image")

mp.subplot(1,2,2)
mp.imshow(sketch,cmap="gray")
mp.axis("off")
mp.title("Final Sketch Image")
mp.show()
```
